Despite its improvements, it still faces challenges such as recognizing complex hand movements, reading data in a way that humans can understand, and recognizing patterns that are unique to people.

##Part B – Compare and evaluate AI techniques. ##

The techniques chosen for this task are Convolutional Neural Networks, Hidden Markov Models, and Transformers. Each technique has unique strengths and is well-suited for addressing different aspects of gesture recognition.

***Convolutional Neural Network (CNN)***

A convolutional neural network is a type of aritifical neural network which is specifically use for image and video recognition. It contains convolution layers which is use to extract features, pooling layers which downsizes the extracted feature maps and fully connected layers which has neurons that is able to learn from the features. CNNs are commonly use in image classification, object detection and image segmentation tasks. Because of the ability to hand large scale images and extract features effectively, this technique is one of the most demending in the industry.

Thre are many pros of this technique. CNNs have high accuracy than other traditional methods. Because of it's ability to detecting and recognizing patterns, this method is ideal for static gesture recognitions. Moreover, this model has automatic feature engineering. It can reduce heavy workload from the designers and increase the efficiency of systems. Also, CNNs can be scale into deeper level. Pre train models such as ResNet prove it by enhancing the accuracy than other CNN models.

The biggest drawback here is that it needs large number of labeled data to achive a high accuracy. Finding a suitable dataset or creating a dataset that can be use to train CNN is harder than other finding datasets for other models. Also, CNNs does not have temporal awareness. It makes harder to capture temporal dynamics from given data.

CNN can be use to detect hand signs from static images. It is ideal for identifying hand signs from image datasets. Also, it can be combine with other models such as LSTM to capture feature from dynamic data. Moreover CNNs can be use to track objects in real time.

These models need large datasets to train and test data efficiently. In most cases, data augmentation need to perfom to scale up the dataset to enhance the accuray of the model. Nowadays there are many datasets that is CNN friendly. But the use of these datasets depends on the perfoming task. Eventhough CNN does not need manual feature extraction, it needs thorough data preparation. It includes steps such as reshaping, scaling and resizing to meet the requirements of the model. Also, finding the most suitable model for the dataset is the most important task to implement a good system.

***Hidden Markov Models (HMM)***

The Hidden Markov Model (HMM) is a statistical model that use to describe relationship between a sequence of observations and a sequence of hidden states. This model has two main variabels, Hidden states: which generate observed data. But these are not directly observable. Observations: variable that is use to meassure and observe. Since, the main concept of Hidden Markov Models is the relationship between hidden states and  observations, there are two main probabilities discuss in this model.


*   Transition probabilities: The probability of moving one hidden leyer to another layer in a signle step.
*   Emission probabilities: The probabilty of observing output given in a specific state in the model.

This model is widely used in classification, clustering and generating tasks. Hnalding complex and non-linear dependencies between hidden states and observations is one of the key advantage of HMM over the other sequential models. Also, these models have more efficieny and scalability than other models. This reduce the computational requiements to implement this model. Also, optimizing HMMs are a possible task. Algorithms such as Viterbi, Baum-Welch and Forward-backward can be use to optimize these models. As well as the interpretability and tranaparent allows to get clear understanding of the hidden states and their transitions. Many developers prefer to use HMM for sequential data because of the above mentioned reasons.

Apart from the advantages, HMMs have several downsides too. One of the most affecting disadvantage is the slower training capability. Depending on the size of the model (hidden states and observations), training these models with large datsets need large computation requirements. Also, the simplistic and restrictive state transitions of the HMMs make strugles for detecting complex gestures. As well as these models need extensive data quality and quantity enhance the accuarcy. Additionaly, these models need a perfect architechture to perfom better without any overfitting or underfitting.

HMMs can be use to detect dynamic gestures such as waves. Also, these models can be use with CNNs for feature extractions. HMMs need sequential datsets for training and testing. Also, the data need extensive preprocessings and scalings. This makes harder when compare to CNNs. By trainig these models, people can detect gesuters in real time and classify gestures or sequences.

***Support Vector Machines (SVM)***

Support Vector Machines (SVMs) are a type of supervised learning models which is mainly use for classification tasks. This model can find the best hyperplane that seperate the classes and maximize the margine between the classes. Maximizing the margine between classes allows the model to classify new data with a better accuracy. These models can handle both linear and non-linear tasks. There are 3 main types of SVMs as Linear SVM, NonLinear SVM and Support vector regression (SVR).

The SVMs can depend with small to medium size datasets. This gave a huge boost to this model to secure a place in the machine learning industry. Also this model does not need an extensive resource allocation when compare to other deep learning models. Also, SVM is one of the best solution to avoid the overfitting because it uses a subset of training make predictions. Also, this models can be regularize. It cause the ability of avoid overfitting too.

Although there are great advantages here, there are also disadvantages. One of the most biggest disadvantage is it is not usually use for large datasets. It is computationaly expensive to use SVM for large datasets. Kernal selection is also a main disadvantage of these models. Kernal selection is directly affect to the model performance. So it is very difficult to use SVM models to train data without knowing deeply about the dataset. Also, the SVM needs great preprocessing of data befor feeding to the model. It can make misclassifications if there are missing values in the dataset.

Training a SVM can be done using small or medium datasets. There are several publicly availabel hand gesture datasets that can be use to train a SVM. In most cases, an SVM can be trained in a short time and requires very clean data. Also, SVMs can make predictions with minimum time. Less training time and fast predictions make this model very popular among the machine learning enthusiasts.

```python
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Reshape the data into 28x28 images
image_data = train_df.iloc[:, 1:].values.reshape(-1, 28, 28, 1)

datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1
)

augmented_images = []
augmented_labels = []

for image, label in zip(image_data, train_df['label']):
    image = np.expand_dims(image, axis=0)

    # Generate augmented images
    for augmented_image in datagen.flow(image, batch_size=1):
        augmented_images.append(augmented_image[0].reshape(28, 28))
        augmented_labels.append(label)

        if len(augmented_images) % 5 == 0:
            break

# Convert augmented data back to DataFrame format
augmented_images_flat = np.array([img.flatten() for img in augmented_images])
augmented_df = pd.DataFrame(augmented_images_flat, columns=train_df.columns[1:])
augmented_df.insert(0, 'label', augmented_labels)

# Combine original and augmented data
train_df_augmented = pd.concat([train_df, augmented_df], ignore_index=True)
```

```python
train_df_augmented.shape
```

```python
#Separating training labels and images
train_images = train_df_augmented.drop('label', axis=1).values
train_labels = train_df_augmented['label'].values
```

```python
#Plot 25 images
class_names = np.unique(train_labels)
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i].reshape(28, 28), cmap='gray')
    plt.xlabel(class_names[train_labels[i]])
plt.show()
```



The test data is a separate df. It will use to test the modeil for final. The training df which has 27455 records will devide into 80/20 ratio and the test set(20% from total) will then devide into 10% , 10% for initial testing and validation.

```python
sns.countplot(x=train_labels)
```

According to the seaborn countplot the dataset have 24 unique labels. Each lable has racords starting from 5500+  to 7500+. We can say this dataset is roughly balanced dataset.

```python
#Extract images from csv file. Need to reshape the images for object recognition task
#The pixels in the csv file is now in 1D array format. So we need to convert it to 2D shape.
images_2D_format = np.array([np.reshape(i, (28, 28)) for i in train_images])
flatten_images = np.array([i.flatten() for i in images_2D_format])
```

```python
#Since the labels are categorical we need to encode it too.
from sklearn.preprocessing import LabelBinarizer
label_binrizer = LabelBinarizer()
labels = label_binrizer.fit_transform(train_labels)
labels
```

Now we will divide training data into 3 phases.
Training data, Testing data and Validation data.

For this we will use Flatten images and encoded labels. First the data will divide into 80% of training data and 20% of tempory data. Then that 20% temp data will divide by 50%. Now the validation images and testing images are 10% from the total dataset.


```python
from sklearn.model_selection import train_test_split

train_images, temp_images, train_labels, temp_labels = train_test_split(
    flatten_images, labels, test_size=0.2, random_state=42
)
```

```python
val_images, test_images, val_labels, test_labels = train_test_split(
    temp_images, temp_labels, test_size=0.5, random_state=42
)
```

```python
# Normalizing image data before feeding to the model.
# First convert images [0, 1] range
x_train = train_images / 255.0
x_val = val_images / 255.0
x_test = test_images / 255.0

#Convert each set into CNN compatible 2D images
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_val = x_val.reshape(x_val.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
```

```python
plt.imshow(x_train[0].reshape(28,28))
```

#Start Building the CNN model

```python
# Defining batch size, epochs and number of classes which is 24
batch_size = 128
num_classes = 24
epochs = 50
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout, GlobalAveragePooling2D
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
# from tensorflow.keras.preprocessing.image import ImageDataGenerator

model = Sequential()
model.add(Conv2D(64, kernel_size=(3,3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(128, kernel_size=(3, 3), activation='relu', kernel_regularizer=l2(0.001)))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(256, kernel_size=(3, 3), activation='relu', kernel_regularizer=l2(0.001)))
model.add(GlobalAveragePooling2D())

model.add(Dense(256, activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(num_classes, activation='softmax'))

optimizer = Adam(learning_rate=1e-3)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history=model.fit(datagen.flow(x_train, train_labels, batch_size=batch_size),
          epochs=epochs,
          validation_data=(x_val, val_labels),
          callbacks=[lr_scheduler])
```
